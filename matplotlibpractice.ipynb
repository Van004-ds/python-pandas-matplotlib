# plotting a single line graph

```python
#1 Plot a line graph showing number of study hours vs scores of 5 students.

import matplotlib.pyplot as plt
plt.rcdefaults()
study_hrs = [2,3,4,5,6]
scores = [20,39,90,23,98]

plt.style.use('seaborn-v0_8-bright')
plt.plot(study_hrs,scores,color = 'hotpink',marker ='o',linewidth=2,linestyle='--',markerfacecolor='yellow',label='performance',markersize=8,markeredgecolor='black') #It's the command that plots your data.Plots a line connecting the (x, y) points
plt.title("Relatioinship between study hrs vs scores",loc ="center",color = "purple",fontsize = 16,fontweight='bold') #Adds a title to your plot
plt.xlabel("study hrs",color ='b',fontsize = 12) #Labels the x-axis
plt.ylabel("score",color ='b',fontsize = 12)# labels the y- axis
plt.grid(True,color ="navy",linestyle =":",linewidth=1) # ads grid lines to the graph for read the data points more clearly
plt.xticks(fontsize=12, color='purple')
plt.yticks(fontsize=12, color='teal')
plt.legend(loc="upper right",fontsize = 12)
ax = plt.gca()
ax.set_facecolor('#fefbd8')
for spine in ax.spines.values():
    spine.set_color("hotpink")

# ax.spines['left'].set_color('hotpink')# only for one side 
plt.figure(figsize=(6,6))
plt.show()






```

```python
import matplotlib.pyplot as plt

# First, set the new style (BEFORE plotting anything!)
plt.rcdefaults()
plt.style.use('Solarize_Light2')

# Then plot
x = [1, 2, 3, 4, 5]
y = [65, 70, 75, 80, 85]

plt.plot(x, y, marker='o', color='lime', linewidth=2)
plt.title("Study Hours vs Score ", fontsize=14, fontweight='bold')
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

# Multiple Line Plots (Comparison Graphs)

```python
#practice sample code ( not a question solution )

hours = [1, 2, 3, 4, 5]
score_student1 = [10, 20, 30, 40, 50]
score_student2 = [12, 18, 28, 38, 45]
plt.plot(hours,score_student1, label = "student 1",color ='hotpink',marker='*',linestyle='--')
plt.plot(hours,score_student2, label = "student 2",color ="teal",marker ="x",linestyle ='-.')
plt.title("sample multiline graph practice graph",color ="blue")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.legend()
plt.figure(figsize=(4,5))
plt.show()



```

```python
#2 . Compare study hours vs scores of 3 students in one graph.
study_hr_week = [ "mon",'tue','wed','thu','fri','sat','sun']
s1_scores =[23,45,65,200,34,45,23]
s2_scores =[2,5,65,20,31,25,82]
s3_scores =[22,39,48,23,39,45,23]
plt.plot(study_hr_week,s1_scores,color="purple", linestyle="--",marker="*")
plt.plot(study_hr_week,s2_scores,color="hotpink", linestyle=':',marker="o")
plt.plot(study_hr_week,s3_scores,color="blue", linestyle='-.',marker="x")
plt.title(" score comparasion of complete week",color ="brown",loc ="center", fontstyle = 'italic')
plt.xlabel("days of the week")
plt.ylabel("scores")
plt.xticks(color ="navy")
plt.yticks(color ="lime")
plt.legend(["student1","student2","student3"])
plt.grid(True)
ax = plt.gca()
ax.set_facecolor("yellow")



```
